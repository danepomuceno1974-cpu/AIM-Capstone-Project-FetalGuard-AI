# *⚗️ **PHASE 07 | CTG — Full Pipeline Demo**

```python
import sys

sys.path.insert(0, "../02 src")

## Full Pipeline in One Script

from data_loader import load_raw
from preprocessor import preprocess
from pipeline import build_pipeline
from model import build_best_model, evaluate, sweep_thresholds, select_optimal_threshold
```

```python
# 1. Load
df_raw = load_raw("../01 data/raw/CTG.xls", sheet_name="Raw Data")
```

```python
# 2. Clean
df_clean, features = preprocess(df_raw)
X = df_clean[features]
y = df_clean["target"]
```

```python
# 3. Split → Scale → Resample
data = build_pipeline(X, y)
```

```python
# 4. Train best model
spw = (data["y_train"] == 0).sum() / (data["y_train"] == 1).sum()
model = build_best_model(spw)
```

```python
# 5. Find optimal threshold
model.fit(data["X_train_scaled"], data["y_train"])
df_thresh = sweep_thresholds(model, data["X_test_scaled"], data["y_test"])
threshold = select_optimal_threshold(df_thresh)
print(f"Optimal threshold: {threshold}")
```

```python
# 6. Final evaluation
results = evaluate(
    model,
    data["X_train_scaled"],
    data["y_train"],
    data["X_test_scaled"],
    data["y_test"],
    threshold=threshold,
    name="XGB (scale_pos_weight)",
)
for k, v in results.items():
    print(f"  {k:<12}: {v}")
```
